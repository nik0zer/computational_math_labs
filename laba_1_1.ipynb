### Условия:
Шаг: $$ h_{n} = \frac{2}{2 ^ n} $$   
Функции: $$sin(x^2),\: cos(sin(x)),\: exp(sin(cos(x))),\: ln(x+3),\: (x+3)^{0.5}$$
Формулы для вычисления:
$$\frac{f(x+h) - f(x)}{h}$$
$$\frac{f(x) - f(x-h)}{h}$$
$$\frac{f(x+h) - f(x-h)}{2h}$$
$$\frac{4}{3}\frac{f(x+h) - f(x-h)}{2h} - \frac{1}{3}\frac{f(x+2h) - f(x-2h)}{4h}$$
$$\frac{3}{2}\frac{f(x+h) - f(x-h)}{2h} - \frac{3}{5}\frac{f(x+2h) - f(x-2h)}{4h} + \frac{1}{10}\frac{f(x+3h) - f(x-3h)}{6h}$$


Функции:

```python
import math
import numpy as np

def func_1(x):
    return np.sin(x ** 2)

def func_2(x):
    return np.cos(np.sin(x))

def func_3(x):
    return np.exp(np.cos(np.sin(x)))

def func_4(x):
    return np.log(x + 3)

def func_5(x):
    return (x + 3) ** 0.5
```

Графики функций:


```python
import matplotlib.pyplot as plt

size = 10.0

x=np.linspace(-1*size, size, 1024)
plt.figure(figsize=[6, 4])
plt.plot(x, [func_1(x_0) for x_0 in x])
plt.tight_layout() 
plt.grid()

x=np.linspace(-1*size, size, 1024)
plt.figure(figsize=[6, 4])
plt.plot(x, [func_2(x_0) for x_0 in x])
plt.tight_layout() 
plt.grid()

x=np.linspace(-1*size, size, 1024)
plt.figure(figsize=[6, 4])
plt.plot(x, [func_3(x_0) for x_0 in x])
plt.tight_layout() 
plt.grid()

x=np.linspace(-1*size, size, 1024)
plt.figure(figsize=[6, 4])
plt.plot(x, [func_4(x_0) for x_0 in x])
plt.tight_layout() 
plt.grid()

x=np.linspace(-1*size, size, 1024)
plt.figure(figsize=[6, 4])
plt.plot(x, [func_5(x_0) for x_0 in x])
plt.tight_layout() 
plt.grid()
```

### Производные функций:
$$2x *cos(x^2),\:-sin(sin(x))cos(x),\:-exp(cos(sin(x)))*sin(sin(x))cos(x),\:\frac{1}{x+3},\:\frac{1}{2\sqrt{x+3}} $$

```python
def func_1_der(x):
    return 2 * x * np.cos(x ** 2)

def func_2_der(x):
    return -1 * np.sin(np.sin(x)) * np.cos(x)

def func_3_der(x):
    return -1 * np.exp(np.cos(np.sin(x))) * np.sin(np.sin(x)) * np.cos(x)

def func_4_der(x):
    return 1 / (x + 3)

def func_5_der(x):
    return 1 / (2 * np.sqrt(x + 3))
```

```python
size = 10.0

x=np.linspace(-1*size, size, 1024)
plt.figure(figsize=[6, 4])
plt.plot(x, [func_1_der(x_0) for x_0 in x])
plt.tight_layout() 
plt.grid()

x=np.linspace(-1*size, size, 1024)
plt.figure(figsize=[6, 4])
plt.plot(x, [func_2_der(x_0) for x_0 in x])
plt.tight_layout() 
plt.grid()

x=np.linspace(-1*size, size, 1024)
plt.figure(figsize=[6, 4])
plt.plot(x, [func_3_der(x_0) for x_0 in x])
plt.tight_layout() 
plt.grid()

x=np.linspace(-1*size, size, 1024)
plt.figure(figsize=[6, 4])
plt.plot(x, [func_4_der(x_0) for x_0 in x])
plt.tight_layout() 
plt.grid()

x=np.linspace(-1*size, size, 1024)
plt.figure(figsize=[6, 4])
plt.plot(x, [func_5_der(x_0) for x_0 in x])
plt.tight_layout() 
plt.grid()
```

### Формулы для погрешностей:

```python
def formule_1(func, x, h):
    return (func(x + h) - func(x)) / h

def formule_2(func, x, h):
    return (func(x) - func(x - h)) / h

def formule_3(func, x, h):
    return (func(x + h) - func(x - h)) / (2 * h)

def formule_4(func, x, h):
    return (4 / 3 * (func(x + h) - func(x - h)) / (2 * h)) - (1 / 3 * (func(x + 2 * h) - func(x - 2 * h)) / (4 * h))

def formule_5(func, x, h):
    return (3 / 2 * (func(x + h) - func(x - h)) / (2 * h)) - (3 / 5 * (func(x + 2 * h) - func(x - 2 * h)) / (4 * h)) + (1 / 10 * (func(x + 3 * h) - func(x - 3 * h)) / (6 * h))
```

### Графики погрешностей:

```python
def my_max(array):
    if(len(array) == 0):
        return np.nan
    max_num = array[0]
    for num in array:
        if(num != np.nan):
            if(max_num == np.nan):
                max_num = num
            else:
                if(num > max_num):
                    max_num = num
    return max_num


def error_rate(func, func_der, formula, h, left_border, right_border, num_of_points):
    x = np.linspace(left_border, right_border, num_of_points)
    return np.abs(my_max(func_der(x) - formula(func, x, h)))

def error_graph(func, func_der, n_start, n_end, num_of_n, left_border, right_border, num_of_points, plot_figsize=[6,4]):
    n = np.linspace(n_start, n_end, num_of_n)
    h = 2 / np.float_power(2, n)
    
    plt.figure(figsize=plot_figsize)
    plt.plot(np.log2(h), [np.log10(error_rate(func, func_der, formule_1, h_current, left_border, right_border, num_of_points)) for h_current in h], 'r')
    plt.plot(np.log2(h), [np.log10(error_rate(func, func_der, formule_2, h_current, left_border, right_border, num_of_points)) for h_current in h], 'b')
    plt.plot(np.log2(h), [np.log10(error_rate(func, func_der, formule_3, h_current, left_border, right_border, num_of_points)) for h_current in h], 'g')
    plt.plot(np.log2(h), [np.log10(error_rate(func, func_der, formule_4, h_current, left_border, right_border, num_of_points)) for h_current in h], 'y')
    plt.plot(np.log2(h), [np.log10(error_rate(func, func_der, formule_5, h_current, left_border, right_border, num_of_points)) for h_current in h], 'k')
    

error_graph(func_1, func_1_der, 1.0, 21.0, 1024, 2, 20, 1024, [20, 10])
error_graph(func_2, func_2_der, 1.0, 21.0, 1024, 2, 20, 1024, [20, 10])
error_graph(func_3, func_3_der, 1.0, 21.0, 1024, 2, 20, 1024, [20, 10])
error_graph(func_4, func_4_der, 1.0, 21.0, 1024, 2, 20, 1024, [20, 10])
error_graph(func_5, func_5_der, 1.0, 21.0, 1024, 2, 20, 1024, [20, 10])
```
